```python
import cv2 as cv
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread('Techno war.jpeg')
#cv.imshow('Techno war', img)
cv2_imshow(img)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

cv2_imshow(gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
classifier = cv.CascadeClassifier('haar_face.xml')
faces_rect = classifier.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)

#print(f'Number of faces found = {len(faces_rect)}')

for (x,y,w,h) in faces_rect:
    cv.rectangle(img, (x,y), (x+w,y+h), (0,255,0), thickness=2)

#plt.imshow(img , cmap="gray")
cv2_imshow(img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
haar_cascade = cv.CascadeClassifier('haar_face.xml')
```

```python
capture = cv.VideoCapture(0)
haar_cascade = cv.CascadeClassifier('haar_face.xml')
counter = 0
while True:
    isTrue, img = capture.read()
    if isTrue:
        faces_rect = haar_cascade.detectMultiScale(img, scaleFactor=1.2, minNeighbors=2)
        for (x,y,w,h) in faces_rect:
            cv.rectangle(img, (x,y), (x+w,y+h), (0,255,0), thickness=2)
        cv.imshow('my video', img)
        if cv.waitKey(20) & 0xFF==ord('x'):
            break
    else:
        break

capture.release()
cv.destroyAllWindows()
```
